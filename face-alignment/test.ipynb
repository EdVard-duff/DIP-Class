```python
import face_alignment
from skimage import io
import numpy as np
import matplotlib.pyplot as plt

fa = face_alignment.FaceAlignment(face_alignment.LandmarksType._2D, flip_input=False,device='cuda:0',face_detector='blazeface')

input = io.imread('..\I2G\man.jpeg')
preds = fa.get_landmarks(input)
preds = np.array(preds)
preds.shape
```

```python
plt.imshow(input)
for detection in preds:
    plt.scatter(detection[:,0],detection[:,1])
```
